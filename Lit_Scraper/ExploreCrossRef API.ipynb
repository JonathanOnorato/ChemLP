Ok, the next step is to go from querying specifically the titles, DOIs, and other metadata, to pulling full articles.

From a planning perspective: 
We have all the metadata, but we really only need the DOI to search for it in the Wiley. I think that means that the smart move from here is to just save the DOIs in the first one to a list, then just move through the list top to bottom, deleting each article from the list as we go. We can use the pop feature to get each one as we go. Start by randomizing it, then pull one off at a time, and find full text. If we have a problem where the function errors out, for example, due to running out of queries, then we should save the list somewhere, so we can start again from where we left off. This will prevent issues of having to check our whole list of dowloadeded articles every time we want to scrape more fulltexts, helping to minimize the problem of starting and stopping when we inevitably run out of downloads for our licenses.

If we want to add more to the list, we can actually just append to that same DOI list. I like this idea very much.

```python
doi_list = ["10.1002/polb.23656", "10.1002/polb.23617", "10.1002/polb.23620"]
```

Next we'll build a method that randomizes our list, then pops off the list. 


```python
doi_list = shuffle(doi_list)
print(doi_list)
#Tell it to pop the item at position 0, top of the list. 
doi = doi_list.pop(0)
print(doi)
```
