The first function <code> selenium_html_collector(driver, url)</code> takes two variables: 1) driver, which is eual to <code> webdriver.the choice of your driver()</code> 2) a url, a book chapter url from  <a herf="https://chem.libretexts.org/Bookshelves" target="_black"> ChemLibretexts</a>.

The second function <code> chapter_text_parser()</code> is responsible for parsing the sub chapter titles, overviews, and contents. And finally setores theses parsed text data to a json file named with the name of the chapter.

```python
import pandas as pd
from lxml import html
import selenium
from bs4 import BeautifulSoup as BS
from selenium import webdriver
import random
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import json
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException

def book_text_parser(url):
    """"This function takes one variable 'url' a book chapter url from Chem.Libretexts""" 
    driver = webdriver.Chrome()
    driver.get(url)
    driver.implicitly_wait(random.randint(1,30))   

    driver.implicitly_wait(random.randint(1,100))
    main_content_container = driver.find_element(By.CLASS_NAME, 'mt-content-container')
    book_link_container = main_content_container.find_elements(By.TAG_NAME, 'a')
    book_links = []
    book_titles = []
    for link in book_link_container:
        if 'Book:' in link.get_attribute('title'):
            book_titles.append(link.get_attribute('title').strip()) 
            book_links.append(link.get_attribute('href')) 
        else:
            pass
    
    driver.get(book_links[9])
    driver.page_source
    book_title = driver.find_element(By.XPATH, '//*[@id="title"]').text.strip()
    main_content_container = driver.find_element(By.CLASS_NAME, 'mt-sortable-listings-container')
    book_chapter_link_container = main_content_container.find_elements(By.TAG_NAME, 'a')
    book_chapter_links = []
    book_chapter_titles = []
    for chapter_link in book_chapter_link_container:
        if 'Front Matter' in chapter_link.get_attribute('title'):
            pass
        elif 'Back Matter' in chapter_link.get_attribute('title'):
            pass
        elif 'Author' in chapter_link.get_attribute('title'):
            pass
        elif 'Solutions' in chapter_link.get_attribute('title'):
            pass
        else:
            book_chapter_links.append(chapter_link.get_attribute('href'))
            book_chapter_titles.append(chapter_link.get_attribute('title'))
   
    subchapter_links =[]
    subchapter_titles =[]
    subchapter_contents=[]
    
    for chapter in book_chapter_links:
        driver.get(chapter)
        driver.page_source
        subchapter_container = driver.find_element(By.CLASS_NAME, 'noindex')
        subchapter_link_container = subchapter_container.find_elements(By.TAG_NAME,'a')
        
        for link in subchapter_link_container:
            try:
                if 'Solutions' in link.get_attribute('title'):
                    pass
                elif 'Problems' in link.get_attribute('title'):
                    pass
                else:
                    subchapter_links.append(link.get_attribute('href'))
                    subchapter_titles.append(link.get_attribute('title'))
            except:
                StaleElementReferenceException, NoSuchElementException
    
    for link in subchapter_links:
        driver.get(link)
        driver.page_source
        subchapter_text_container = driver.find_elements(By.CLASS_NAME,'mt-content-container') 
        for subchap in subchapter_text_container:
            subchapter_contents.append(subchap.text.strip())
         
            
    data = {'chapter title': book_chapter_titles,'subchapter title': subchapter_titles, 'subchapter content': subchapter_contents}
    
    filename = book_title
    
    with open(filename, "w") as outfile:
            json.dump(data, outfile)
            
    return print(len(book_titles)), subchapter_titles
```
