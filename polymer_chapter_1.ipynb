```python
import pandas as pd
import re
from bs4 import BeautifulSoup as BS
import requests
import numpy as np
from lxml import html
import selenium
from selenium import webdriver
import time
import os
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import itertools
import json
```

```python
url = 'https://chem.libretexts.org/Bookshelves/Organic_Chemistry/Book%3A_Polymer_Chemistry_(Schaller)/1%3A_Monomers_and_Polymers'

driver = webdriver.Chrome()
driver.get(url)
driver.implicitly_wait(100)


#prints titles of sub_chapters from home page url

soup = BS(driver.page_source, 'html.parser')
chapter_link = soup.find_all('dl', {'class': 'mt-listing-detailed'})
chapter_titles =[title.find('a')['title'] for title in chapter_link]
print('chapter titles :', '\n',chapter_titles)

# scrapes all the sub_chapters 

i = 0
chapter_links = driver.find_elements(By.CLASS_NAME, 'mt-icon-next-article')

while i < len(chapter_links):
    try:
        driver.find_element(By.CLASS_NAME, 'mt-icon-next-article').click()
        # driver cilicks next article
    except:
        continue
    
    # Beautifulsoup gets page source from driver
    
    soup = BS(driver.page_source, 'html.parser')
    
    #prints subchapter titles
    
    subchapter_title = [t.find('span') for t in soup.find_all('a', {'class':'mt-icon-next-article'})]
    print('\n\n\n' , subchapter_title)
    
    #gets chapter_content_text
    
    whitelist = ['p']
    chapter_text = [t for t in soup.find_all(text=True) if t.parent.name in whitelist]
    print('\n\n',chapter_text)
    
    #gets chapter_content_image
    
    chapter_image = soup.find_all('p',{'class':'mt-align-center'})
    image = [i.find('img')['src'] for i in chapter_image]
    print('\n', '\n',image)
    
     # the driver is set to top at subchapter 1.6. Because the above code to scrape image became insubscriptable. I tired to find more universal code but have not yet succeed.
        
    if "1.6: Supramolecular Assemblies" in driver.find_element(By.CLASS_NAME, 'mt-icon-next-article').get_attribute("title"):
        break
   

# saves scraped data to a json file

polymer_chapter1_data = {'subchapter-titles': 'subchapter_title', 'chaptercontet-text': 'chapter_text', 'chapter-iamge': 'image'}

json_file = open('out.json', 'w')
json.dump(polymer_chapter1_data, json_file)
json_file.close()

driver.close()
```
