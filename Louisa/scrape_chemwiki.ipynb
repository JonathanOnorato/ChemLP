```python
from time import sleep
from random import randint
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as soup
```

```python
#uses chromedriver to open chrome via Selenium
chromedriver_path = 'C:/Users/louis/Downloads/chromedriver_win32/chromedriver.exe'
#adds a delay in the execution of opening chrome by a random number of seconds between 1 and 30
sleep(randint(1, 30))

driver = webdriver.Chrome(chromedriver_path)

url = 'https://www2.chemistry.msu.edu/faculty/reusch/VirtTxtJml/intro2.htm#start'
#waits a randome number of seconds between 1 and 30 before opening the webpage
driver.implicitly_wait(randint(1, 30))
#literally getting the webpage
driver.get(url)
```

```python
#Selenium gives BeautifulSoup the job for parsing
page_soup = soup(driver.page_source, 'html')

content = page_soup.get_text()

print(content)
```
